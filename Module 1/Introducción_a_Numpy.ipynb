#Difference between Python list and Numpy array

Python List
*   Can contain diffent data types in a single list
* Do no support broadcast
* Lists can dynamically grow and shrink in size

Numpy Array
*   All elements should be homogeneous.
*   The lists are slower for numerical computations because they involve interpreted Python code
* Array consumes less memory, they store data in a contiguous block of memory
* The arrays are optiimized for numerical operations (they use compiled C code)

* Provides a wide range of functions and methods for mathematical operations:**broadcast, slicing, reshaping and aggregations**.
* Numpy arrays have fixed size and shape once is created.
