```python
# Importar las librer√≠as necesarias
import pandas as pd
import os
```

```python
# Descargar el dataset "Titanic" desde una URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Obtener informaci√≥n general sobre el DataFrame
df.info()
```

```python
# Describir las estad√≠sticas b√°sicas de las columnas num√©ricas
df.describe()
```

```python
# Seleccionar una columna espec√≠fica (por ejemplo, la edad)
age_column = df['Age']
age_column.head()
```

```python
# Filtrar el DataFrame para mostrar solo pasajeros mayores de 30 a√±os
filtered_df = df[df['Age'] > 30]
filtered_df.head()
```

```python
# A√±adir una nueva columna basada en una operaci√≥n (por ejemplo, un flag para menores de edad)
df['IsMinor'] = df['Age'] < 25
df.head()
```

```python
# Eliminar filas con valores faltantes (NaN)
df_cleaned = df.dropna()
df_cleaned.head()
```

```python
# Guardar el DataFrame modificado en un nuevo archivo CSV
df_cleaned.to_csv('cleaned_titanic_data.csv', index=False)
```
